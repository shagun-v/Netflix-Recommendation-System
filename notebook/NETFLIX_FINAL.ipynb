```python
def parse_netflix_file(filepath):

    movie_ids = []
    user_ids = []
    ratings = []
    dates = []

    current_movie = None

    with open(filepath, "r", encoding="latin1") as f:

        for line in tqdm(f):

            line = line.strip()

            if ":" in line:

                current_movie = int(line[:-1])

            else:

                user, rating, date = line.split(",")

                movie_ids.append(current_movie)
                user_ids.append(int(user))
                ratings.append(int(rating))
                dates.append(date)

    return pd.DataFrame({
        "User_ID": user_ids,
        "Movie_ID": movie_ids,
        "Rating": ratings,
        "Date": dates
    })
```

# MODEL 2 — ITEM BASED COLLABORATIVE FILTERING

Item-Based Collaborative Filtering recommends movies by identifying similar movies based on historical rating behavior.

Cosine similarity is used to measure similarity between movie vectors.

```python
from surprise import Dataset
from surprise import Reader

from surprise import KNNBasic

from surprise import accuracy

from surprise.model_selection import train_test_split
```

```python
ratings_sample = ratings.sample(
    n=2000000,
    random_state=42
)

print(ratings_sample.shape)
```

```python
ratings_sample.head()
```

```python
print(
    "Unique Users:",
    ratings_sample["User_ID"].nunique()
)

print(
    "Unique Movies:",
    ratings_sample["Movie_ID"].nunique()
)
```

```python
reader = Reader(
    rating_scale=(1,5)
)

data = Dataset.load_from_df(
    ratings_sample[
        ["User_ID","Movie_ID","Rating"]
    ],
    reader
)
```

```python
trainset, testset = train_test_split(
    data,
    test_size=0.2,
    random_state=42
)
```

```python
sim_options = {
    "name": "cosine",
    "user_based": False
}

item_cf = KNNBasic(
    sim_options=sim_options
)

item_cf.fit(trainset)
```

```python
predictions = item_cf.test(
    testset
)
```

```python
item_cf_rmse = accuracy.rmse(
    predictions,
    verbose=False
)

print(
    f"Item-CF RMSE: {item_cf_rmse:.4f}"
)
```

# MODEL 3 — MATRIX FACTORIZATION (SVD)

SVD decomposes the user-item interaction matrix into latent user and movie factors.

This approach captures hidden preference patterns and generally performs better than neighborhood-based methods on sparse datasets.

```python
from surprise import SVD
```

```python
svd = SVD(
    n_factors=100,
    n_epochs=20,
    lr_all=0.005,
    reg_all=0.02,
    random_state=42
)
```

```python
svd.fit(trainset)
```

```python
svd_predictions = svd.test(
    testset
)
```

```python
svd_rmse = accuracy.rmse(
    svd_predictions,
    verbose=False
)

print(
    f"SVD RMSE: {svd_rmse:.4f}"
)
```

```python
from collections import defaultdict
import numpy as np
```

```python
def get_top_n(predictions, n=10):

    top_n = defaultdict(list)

    for uid, iid, true_r, est, _ in predictions:

        top_n[uid].append((iid, est, true_r))

    for uid, user_ratings in top_n.items():

        user_ratings.sort(
            key=lambda x: x[1],
            reverse=True
        )

        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
def mapk(predictions, k=10, threshold=3.5):

    top_n = get_top_n(predictions, n=k)

    APs = []

    for uid, user_ratings in top_n.items():

        num_hits = 0
        score = 0

        for i, (_, _, true_rating) in enumerate(user_ratings):

            if true_rating >= threshold:

                num_hits += 1

                score += num_hits / (i + 1)

        if num_hits > 0:

            APs.append(score / num_hits)

    return np.mean(APs)
```

```python
item_cf_map10 = mapk(
    predictions,
    k=10
)

print(
    f"Item-CF MAP@10: {item_cf_map10:.4f}"
)
```

```python
svd_map10 = mapk(
    svd_predictions,
    k=10
)

print(
    f"SVD MAP@10: {svd_map10:.4f}"
)
```

```python
comparison = pd.DataFrame({
    "Model":[
        "Item-Based CF",
        "SVD"
    ],
    "RMSE":[
        round(item_cf_rmse,4),
        round(svd_rmse,4)
    ],
    "MAP@10":[
        round(item_cf_map10,4),
        round(svd_map10,4)
    ]
})

comparison
```

```python
def precision_at_k(predictions,
                   k=10,
                   threshold=3.5):

    top_n = get_top_n(predictions, n=k)

    precisions = []

    for uid, user_ratings in top_n.items():

        n_rel = sum(
            (true_r >= threshold)
            for (_, _, true_r)
            in user_ratings
        )

        precisions.append(
            n_rel / k
        )

    return np.mean(precisions)
```

```python
item_cf_precision = precision_at_k(
    predictions
)

svd_precision = precision_at_k(
    svd_predictions
)

print(item_cf_precision)
print(svd_precision)
```

```python
import pandas as pd

#Recommendation Function
def recommend_movies(user_id,
                     model,
                     ratings_df,
                     movie_titles_df,
                     top_n=10):

    watched = set(
        ratings_df[
            ratings_df["User_ID"] == user_id
        ]["Movie_ID"]
    )

    movie_support = ratings_df["Movie_ID"].value_counts()

    all_movies = movie_support[
        movie_support >= 500
    ].index



    unseen_movies = [
        movie
        for movie in all_movies
        if movie not in watched
    ]

    predictions = []

    for movie_id in unseen_movies:

        pred = model.predict(
            user_id,
            movie_id
        )

        predictions.append(
            (
                movie_id,
                float(pred.est)
            )
        )

    predictions.sort(
        key=lambda x: x[1],
        reverse=True
    )

    top_predictions = predictions[:top_n]

    recommendations = []

    for movie_id, score in top_predictions:

        title = movie_titles_df[
            movie_titles_df["Movie_ID"] == movie_id
        ]["Title"].values

        title = title[0] if len(title) > 0 else "Unknown"

        recommendations.append(
            (
                movie_id,
                title,
                float(score)
            )
        )

    return pd.DataFrame(
        recommendations,
        columns=[
            "Movie_ID",
            "Title",
            "Predicted_Rating"
        ]
    )
```

## Failure Cases

Although SVD achieved strong RMSE and MAP@10 scores, collaborative filtering methods may fail when:

- Users provide very few ratings.
- Movies have limited interaction history.
- User preferences change over time.

These situations reduce recommendation quality and highlight the need for hybrid recommendation approaches.

### Cold Start Problem

New User Problem:
The system cannot generate personalized recommendations for users with no historical interactions.

New Item Problem:
Movies with very few ratings have insufficient interaction data for collaborative filtering models.

Possible Solutions:
- Hybrid recommendation systems
- Content-based filtering
- Popularity-based fallback recommendations
